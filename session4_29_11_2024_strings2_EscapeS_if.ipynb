1- input() ile kullanıcıdan yaşını öğrenin

2- Eğer yaşı 30'un altındaysa "Çok gençsin" yazdırın

3- Eğer yaşı 31 (dahil) - 60(dahil) aralığındaysa "Olgunlaşmışsın"

4- Eğer yaşı 60'dan büyükse "   "

```python
yas = int(input("Yaşınızı giriniz"))
if 0 <= yas <= 30:
    print("Gençsiniz")
elif yas >= 31 and yas <= 60:  # 31 <= yas <= 60
    print("Olgunlaşmışsınız")
elif 100 >= yas > 60:
    print("Bilge kişilik")
else:
    print("Hatalı giriş")  
```

## nested if

```python
if 3 < 9:
    if 5 > 8:
        print("This is true")
    else:
        print("This is false")
elif 4 > 9:
    print("This is else")
```

Kullanıcıdan input() ile notunu öğrenin. Harf sistemine göre bu notun harf karşılığını veren python programını yazın.

```python
x = int(input("Lütfen puaninizi giriniz:"))
if x >= 90:
    notun = 'A'
elif x >= 80:
    notun = 'B'
elif x >= 70:
    notun = 'C'
elif x >= 60:
    notun = 'D'
else:
    notun = 'E'
print(f"Girdiginiz puan: {x}, Harf notunuz: {notun}")
```
