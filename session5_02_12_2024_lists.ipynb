Kullanıcıdan bir sayı alınız ve sayının pozitif, negatif veya sıfır olup olmadığını ekrana yazdırınız.

```python
sayi = int(input("Bir sayı giriniz."))
if sayi < 0:
    print("Girilen sayı negatif.")
elif sayi > 0:
    print("Girilen sayı pozitif.")
else:
    print("Girilen sayı sıfır.")
```

Kullanıcıdan alınan sayının tek mi çift sayı mı olduğunu yazdırınız.

```python
sayi = int(input("Bir sayı giriniz."))
if sayi % 2 == 0:
    print("Girdiğiniz sayı çifttir.")
else:
    print("Girdiğiniz sayı tekttir.")
```

Kullanıcıdan bir karakter alınız ve karakterin büyük harf, küçük harf veya rakam olup olmadığını kontrol ediniz.

```python
karakter = input("Bir karakter giriniz.")
if karakter.isupper():
    print("Büyük harf.")
elif karakter.islower():
    print("Küçük harf.")
elif karakter.isnumeric():
    print("Rakam")
else:
    print("Diğer karakter") 
```

Kullanıcıdan bir yıl alınız ve yılın artık yıl olup olmadığını kontrol ediniz. (Artık yıllar 4'e tam bölünen, ancak 100'e tam bölünemeyen veya 400'e tam bölünen yıllardır.)

```python
yil = int(input("Bir yıl giriniz."))
if (yil % 4 == 0 and yil % 100 != 0 or yil % 400 == 0):
    print(f"{yil} bir artık yıldır.")
else:
    print(f"{yil} bir artık yıl değildir.")
```
